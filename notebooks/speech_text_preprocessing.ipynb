```python
from glob import glob

v1_dfs = np.sort(np.array(glob('./*v1.csv')))
v1_dfs[0].strip("./")
```

```python
!pwd
```

```python
for d in v1_dfs:
    print(d.strip('./'))
```

```python
df_name = 'youtube_data.csv'

for idx, d in enumerate( v1_dfs ):
    if idx != 0:

        df0 = pd.read_csv( df_name )
        df1 = pd.read_csv( d.strip("./") )

        pd.concat( [df0, df1], ignore_index=True ).to_csv( df_name, index=False )

    else:
        df = pd.read_csv( d.strip("./") )
        df.to_csv( df_name, index=False )

print( "All done" )
```
